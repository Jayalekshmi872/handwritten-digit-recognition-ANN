###basic AI Project

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(x_train,y_train),(x_test,y_test) = keras.datasets.mnist.load_data()
```

```python
x_train.shape    #60,000 images for training
```

```python
x_test.shape    #10000 for test
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[0])   #for displaying the first images in x_train
# this is an unstructured data
# this is an grey scaled image (0 to 255)
```

```python
y_train[0] #corresponding answer
```

```python
x_train[0]   #values range from 0 to 255
```

```python
# if the values are in different range we need to scale them
#for scale divide each with 255 (max range) and we get number range from 0 to 1
```

```python
x_train = x_train/255
```

```python
x_train[0]
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[0])   #no change to the image
```

```python
x_test = x_test/255
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_test[0])
```

###ANN

```python
#create a sequential ANN
model = Sequential()

# we have 3 layer [input, hidden, output]
# first is input, here input is 28*28 image
#we need to add a layer ,and that layer should be flatten
model.add(Flatten(input_shape = (28,28))) #input layer
#after flatten 28*28 = 784

#hidden layer -two
model.add(Dense(128,activation='relu'))
#128 - neurons
model.add(Dense(32,activation='relu'))
#32 - neurons

#output layer
model.add(Dense(10,activation='softmax'))
#10 neurons
#only 10 we can set bcoz the number is 0-9 total 10 number

#we can decide how many neurons to set for hidden layer
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
#loss function is to minimize the error  b/w the actual and predict output
```

```python
model.fit(x_train,y_train,epochs=50,validation_split=0.2)
#epochs ---> ethra iteration il ethu nadakanm
#validation_split ----> 20 percentage data for testing
```

```python
y_prob = model.predict(x_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
